```python
# Почти как надо, но тут для всех данных получается


# Create a list of the mean scores for each variable
mean_values = [df['overcome'].mean(), df['pleased'].mean(), df['enjoying'].mean()]
# Create a list of variances, which are set at .25 above and below the score
variance = [df['overcome'].mean() * 0.25, df['pleased'].mean() * 0.25, df['enjoying'].mean() * 0.25]
# Set the bar labels
bar_labels = ['Overcome', 'Pleased', 'Enjoying']
# Create the x position of the bars
x_pos = list(range(len(bar_labels)))
# Create the plot bars
# In x position
plt.bar(x_pos,
        # using the data from the mean_values
        mean_values, 
        # with a y-error lines set at variance
        yerr=variance, 
        # aligned in the center
        align='center',
        # with color
        color='#FFC222',
        # alpha 0.5
        alpha=0.5)
# add a grid
plt.grid()

# set height of the y-axis
max_y = max(zip(mean_values, variance)) # returns a tuple, here: (3, 5)
plt.ylim([0, (max_y[0] + max_y[1]) * 1.1])
# set axes labels and title
plt.ylabel('Correct unswers')
plt.xticks(x_pos, bar_labels)
plt.title('Mean Scores For Each Emotion')

plt.show()
```

```python
#Какая то хренотень здесь в итоге получается

# Setting the positions and width for the bars
pos = list(range(len(vaso['overcome']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with overcome data,
# in position pos,
plt.bar(pos, 
        #using df['overcome] data,
        vaso['overcome'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=vaso['group'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using vaso['pleased'] data,
        vaso['pleased'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=vaso['group'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using vaso['enjoying'] data,
        vaso['enjoying'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=vaso['group'][2]) 

# Set the y axis label
ax.set_ylabel('Correct')

# Set the chart's title
ax.set_title('Emotions correct responces')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(vaso['group'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(vaso['overcome'] + vaso['pleased'] + vaso['enjoying'])] )

# Adding the legend and showing the plot
plt.legend(['Overcome', 'Pleased', 'Enjoying'], loc='upper left')
plt.grid()
plt.show()
```
